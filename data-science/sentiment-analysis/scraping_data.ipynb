```python
import requests
import csv
from bs4 import BeautifulSoup
```

# Scraping Dataset

```python
# url
urls = [
    'https://reviews.femaledaily.com/products/makeup-remover/face/pond-s/brightening-micellar-miracle-water-all-in-one',
    'https://reviews.femaledaily.com/products/makeup-remover/face/pond-s/oil-control-micellar-miracle-water-all-in-one'
]
scrapreview_list = []
max_review = 10000
total_review = 0

for product_url in urls:
  page = 1
  while total_review < max_review:
    response = requests.get(f"{product_url}?cat=&cat_id=0&age_range=&skin_type=&skin_tone=&skin_undertone=&hair_texture=&hair_type=&order=newest&page={page}")

    # Check if the request was successful
    if response.status_code == 200:
      print(f"Page {page} successfully fetched!")
      html_content = response.text
    else:
      print(f"Failed to fetch the page {page}")
      html_content = ''

    # Parsing the HTML content with BeautifulSoup
    bsoup = BeautifulSoup(html_content, 'html.parser')

    # Find all review elements
    scrapreview = bsoup.find_all('div', class_='review-card')

    if not scrapreview:
      break

    for review in scrapreview:

      review_list = {}

      review_list['Id'] = total_review + 1
      review_list['Username'] = review.find('p', class_='profile-username').text.strip() if review.find('p', class_='profile-username') else "N/A"
      review_list['Age'] = review.find('p', class_='profile-age').text.strip() if review.find('p', class_='profile-age') else "N/A"
      review_list['Skin_Description'] = review.find('p', class_='profile-description').text.strip() if review.find('p', class_='profile-description') else "N/A"
      review_list['Date'] = review.find('p', class_='review-date').text.strip() if review.find('p', class_='review-date') else "N/A"
      review_list['Rating'] = len(review.find_all('i', class_='icon-ic_big_star_full'))
      review_list['Review'] = review.find('p', class_='text-content').text.strip() if review.find('p', class_='text-content') else "N/A"

      scrapreview_list.append(review_list)
      total_review += 1

    page += 1
```

## Save data

```python
# Write the scraped data into a CSV file
with open('review_product.csv', mode='w', newline='', encoding='utf-8') as file:
  writer = csv.DictWriter(file, fieldnames=['Id', 'Username', 'Age', 'Skin_Description', 'Date', 'Rating', 'Review'])
  writer.writeheader()
  writer.writerows(scrapreview_list)
```

```python
print(f"Extracted {len(scrapreview_list)} reviews and saved them to 'review_product.csv'.")
```
