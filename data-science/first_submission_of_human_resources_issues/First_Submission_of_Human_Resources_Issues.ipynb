```python
# Get the RFM column
rfm_df = main_df[['EmployeeId', 'YearsSinceLastPromotion', 'TrainingTimesLastYear', 'MonthlyIncome']].copy()

# Rename column
rfm_df.columns = ['EmployeeId', 'Recency', 'Frequency', 'Monetary']

# Normalization
scaler = MinMaxScaler()
rfm_df[['R_score', 'F_score', 'M_score']] = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])

# RFM score
rfm_df['RFM_score'] = rfm_df['R_score'] + rfm_df['F_score'] + rfm_df['M_score']
```

```python
rfm_df.head()
```

**Conclusion**
* Recency shows how long ago the employee was promoted. If the score is high, it means they havenâ€™t been promoted in a long time, which may lead to feeling stuck and wanting to leave.

* Frequency tells us how often the employee joins training or company activities. A low score means they donâ€™t join often, which may show they are less active or less involved.

* Monetary means how much the employee earns. A low score means lower salary, which could be a reason to leave the company.

```python
# Calculate quartile limits
q90 = rfm_df['RFM_score'].quantile(0.90)
q75 = rfm_df['RFM_score'].quantile(0.75)
q25 = rfm_df['RFM_score'].quantile(0.25)

# Create segment conditions and options
conditions = [
    rfm_df['RFM_score'] >= q90,
    rfm_df['RFM_score'] >= q75,
    rfm_df['RFM_score'] >= q25
]

options = [
    "High value",
    "Medium value",
    "Low value"
]

rfm_df['Segment'] = np.select(conditions, options, default="Very low value")

rfm_df[['EmployeeId', 'RFM_score', 'Segment']].head()
```

## Modeling

```python
kmeans_df = result_df[['EmployeeId', 'Recency', 'Frequency', 'Monetary', 'Attrition', 'Age']].copy()
kmeans_df.head()
```

```python
def power_transforms(features, df):
  df = df[features]
  for feature in features:
    transformer = PowerTransformer(standardize=True)
    y = np.asanyarray(df[feature])
    y = y.reshape(-1,1)
    transformer.fit(y)
    df["transform_{}".format(feature)] = transformer.transform(y)
    df.drop([feature], axis=1, inplace=True)
    joblib.dump(transformer, "transformer_{}.joblib".format(feature))
  return df
```

```python
features = ['Recency', 'Frequency', 'Monetary', 'Attrition', 'Age']
transformed_kmeans_df = power_transforms(features, kmeans_df)
transformed_kmeans_df.head()
```

```python
num_dis_plot(
    features=["transform_Recency", "transform_Frequency", "transform_Monetary", "transform_Attrition", "transform_Age"],
    df=transformed_kmeans_df,
)
```

```python
inertia = {}
silhouette = {}

# Looping to find inertia and silhouette values
for k in range(2, 10):
    model = KMeans(n_clusters=k, random_state=75)
    model.fit_predict(transformed_kmeans_df)
    inertia[k] = model.inertia_
    silhouette[k] = silhouette_score(transformed_kmeans_df, model.labels_)

# Show the best k value based on the highest silhouette
best_k_silhouette = max(silhouette, key=silhouette.get)
print(f"Best k based on silhouette score: {best_k_silhouette} (Score: {silhouette[best_k_silhouette]:.4f})")

# Show the best k value based on the lowest inertia
best_k_inertia = min(inertia, key=inertia.get)
print(f"Best k based on lowest inertia: {best_k_inertia} (Inertia: {inertia[best_k_inertia]:.4f})")

fig, axs = plt.subplots(1, 2, figsize=(11, 4))

axs[0].plot(inertia.keys(), inertia.values(), marker='o', lw=3)
axs[0].set_xlabel('Number of Clusters', fontsize=12)
axs[0].set_ylabel('Inertia')
axs[0].set_title('The Elbow Method', fontsize=15)

axs[1].plot(silhouette.keys(), silhouette.values(), marker='*', lw=3)
axs[1].set_xlabel('Number of Clusters', fontsize=12)
axs[1].set_ylabel('Silhouette Coefficient')
axs[1].set_title('The Silhouette Method', fontsize=15)

plt.tight_layout()
plt.show()
```

```python
K = 2

model = KMeans(n_clusters=K, random_state=75)
model.fit(transformed_kmeans_df)

joblib.dump(model, "kmeans_clustering_model.joblib")
```

```python
clusters = model.predict(transformed_kmeans_df)

kmeans_df["EmployeeSegment"] = clusters.astype(str)
kmeans_df.sample(5)
```

```python
result_kmeans_df = pd.merge(
    left=main_df,
    right=kmeans_df[["EmployeeId","Recency", "Frequency", "Monetary", "EmployeeSegment"]],
    on="EmployeeId",
    how="inner"
)
result_kmeans_df.head()
```
